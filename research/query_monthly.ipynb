```python
import os
from supabase import create_client, Client
import dotenv
dotenv.load_dotenv()

url: str = os.environ.get("URL")
key: str = os.environ.get("ANON_KEY")
supabase: Client = create_client(url, key)
```

```python
# Using the RPC function
response = (
    supabase.rpc(
        "get_monthly_h3_counts", {"target_month": "2025-07-01"}
    )
    .execute()
)

data = response.data
print(f"Found {len(data)} H3 cells")
for row in data:
    print(f"Cell: {row['h3_cell_start']}, Count: {row['total_count']}")
```

```python
import pandas as pd
total_departures_by_cell_df = pd.DataFrame(data)
```

```python
import geopandas as gpd
gdf_hexagons = gpd.read_file("./geo/nyc__jc_h3_hexagons_9")

gdf_hexagons_w_counts = gdf_hexagons.merge(total_departures_by_cell_df, left_on='h3_id', right_on='h3_cell_end', how='left')

```
